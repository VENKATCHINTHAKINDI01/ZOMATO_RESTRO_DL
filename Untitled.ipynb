```python
import re
import nltk
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

def clean_text(text):
    if pd.isnull(text):
        return ""

    # Lowercase
    text = text.lower()

    # Remove numbers & special characters
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # Tokenize
    tokens = text.split()

    # Remove stopwords + stem
    tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]

    return " ".join(tokens)
```

```python
df['clean_text'] = df['Review'].astype(str).apply(clean_text)
```

```python
from gensim.models import Word2Vec

sentences = df['clean_text'].str.split().tolist()

w2v_model = Word2Vec(
    sentences,
    vector_size=100,     # embedding dimension
    window=5,
    min_count=1,
    workers=4,
    sg=1                 # skip-gram = better semantic learning
)
```

```python
import numpy as np

def get_w2v_vector(tokens, model, size=100):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(size)
    return np.mean(vectors, axis=0)

df['w2v_vector'] = df['clean_text'].str.split().apply(
    lambda x: get_w2v_vector(x, w2v_model)
)
```

```python
import numpy as np

def get_w2v_vector(tokens, model, size=100):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(size)
    return np.mean(vectors, axis=0)

df['w2v_vector'] = df['clean_text'].str.split().apply(
    lambda x: get_w2v_vector(x, w2v_model)
)
```

```python
# Create restaurant-level dataset
restaurant_df = df.groupby("Name").agg({
    "w2v_vector": "first",
    "Rating": "mean",
    "Cuisines": "first",
    "Cost": "first"
}).reset_index()

X = np.vstack(df["w2v_vector"].values)
y = df["Rating"].values
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(128, activation='relu', input_shape=(X.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')  # predicted rating score
])

model.compile(
    optimizer='adam',
    loss='mse'
)

history = model.fit(
    X, y,
    epochs=15,
    batch_size=32,
    validation_split=0.1,
    verbose=1
)
```

```python
def filter_by_cuisine(cuisine):
    # case-insensitive match
    mask = df['Cuisines'].str.contains(cuisine, case=False, na=False)
    return df[mask]
```

```python
def get_restaurant_level_data(df):
    restaurant_df = df.groupby("Name").agg({
        "w2v_vector": "first",
        "Rating": "mean",
        "Cuisines": "first",
        "Cost": "first"
    }).reset_index()

    # Predict score using DL model
    X_embeds = np.vstack(restaurant_df["w2v_vector"].values)
    restaurant_df["predicted_score"] = model.predict(X_embeds).flatten()

    return restaurant_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def compute_similarity_matrix(restaurant_df):
    embed_matrix = np.vstack(restaurant_df["w2v_vector"].values)
    return cosine_similarity(embed_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_top_for_cuisine(cuisine, top_n=10, alpha=0.7):

    # Step 1  filter restaurants by cuisine
    df_cuisine = filter_by_cuisine(cuisine)

    if df_cuisine.empty:
        return f"No restaurants found for cuisine: {cuisine}"

    # Step 2  restaurant-level embedding + DL prediction
    rest_df = get_restaurant_level_data(df_cuisine)

    # Step 3  semantic similarity within cuisine
    embed_matrix = np.vstack(rest_df["w2v_vector"].values)
    sim_matrix = cosine_similarity(embed_matrix)

    # Average similarity per restaurant
    semantic_scores = sim_matrix.mean(axis=1)

    # Hybrid score
    rest_df["hybrid_score"] = (
        alpha * rest_df["predicted_score"] +
        (1 - alpha) * semantic_scores
    )

    # Step 4  return sorted results
    return rest_df.sort_values(by="hybrid_score", ascending=False).head(top_n)[
        ["Name", "Cuisines", "Cost", "Rating", "predicted_score", "hybrid_score"]
    ]
```
